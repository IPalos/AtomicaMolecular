4. El átomo de manganeso (Z=25) tiene el estado base una subcapa exactamente llena a la mitad con 5 electrones. Dé la configuración electrónica y el estado base del átomo.

### Solución
1. Primero obtengamos los posibles niveles debido a la interacción spin-spin. Considerando que tenemos 2 electrones en $1s$ y otros 2 electrones en $2s$, entonces el spin de los orbitales internos es $S=0$, sólo queda considerar el spin de los electrones en $2p$ y $3d$, si son paralelos, tenemos $S=1$, y en caso contrario, tenemos que $S=0$. Ahora considerando la interacción órbita-órbita, si $L_{2_p}=1$ y $L_{3_d}=2$, entonces $L_t=1,2,3$. Por lo tanto hay 3 niveles en el caso de tener spin paralelo, y otros 3 niveles en caso contrario, luego, podemos concluir que hay <b>6 niveles</b> posibles con la configuación dada (${}^1P,{}^1D,{}^1F,{}^3P,{}^3D \; y \;{}^3F$) .

2. Siguiendo las reglas de Hund y el principio de Pauli, la configuración es ${}^4F_{3/2}$, con $L=3, \;S=3/2, \;J=3/2 $, entonces

\begin{align}
\left|\left(\mu_{J}\right)_{J}\right|&=\frac{3 J(J+1)+S(S+1)-L(L+1)}{2 \sqrt{J(J+1)}} \cdot \mu_{\mathrm{B}} \\
{}&= 0.77\cdot\mu_B
\end{align}

3. Para el caso de $4d5s^2$, su configuración es ${}^2D_{3/2}$ y para el caso de $4d^25s^2$, su configuración es ${}^3F_{2}$

4. La configuración electrónica es $[Ar]3d^54s^2$  y su configuración atómica en el estado base es ${}^6S_{5/2}$

# 9. <span id="Problema9"> Muestre que la suma $\sum(2J+1)$ para valores dados de L y S es igual al producto $(2L+1)(2S+1)$ ¿ Cuál es el significado físico de este producto?

## Solución

Sea $a$ tal que $S=\frac{a}{2}$, entonces 

\begin{align}
\sum_{i=0}^a (2J+1) &= \sum_{i=0}^a 2\left(L+\frac{a}{2}-i\right)+1 \\
{}&= ((2L+a+1)+(2L-a+1))+((2L+a-2+1)+(2L-a+2+1))+... \\
{}&= \underbrace{2(2L+1)}_{\frac{a+1}{2} \text{veces}} \\
{}&= (2S+1)(2L+1)
\end{align}

Ese producto es la cantidad de microestados que contiene el estado ${}^{(2S+1)}L_J$
